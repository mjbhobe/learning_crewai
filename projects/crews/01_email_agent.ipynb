# 01 Email Agent
In this notebook, we'll illustrate how we can use a simple CrewAI powered agent to write professional emails from raw text.

We'll be using Google Gemini 3.1 Flash LLM. Install the following packages in your local environment:
```bash
$> uv acc crewai[tools] crewai[google-genai]
```

````python
from dotenv import load_dotenv
from rich.console import Console
from rich.markdown import Markdown

load_dotenv(override=True)
console = Console()
````

````python
from crewai.llm import LLM

llm = LLM(
    model="gemini/gemini-3.1-flash-lite",
    temperature=0.1,
)

response = llm.call("Who invented the Electric bulb?")
print(response)
````

````python
from crewai import Agent, Task, Crew
import asyncio

email_agent = Agent(
    # a simplr description of the agent
    role="email assistant",
    # what the agent should do
    goal="Improve emails and make them sound professional and clear",
    # what role should agent play "You are an expert Financial Analyst..."
    backstory="A highly expert communication specialist skilled in professional email writing",
    llm=llm,
    # show all calls
    verbose=True,
)

original_email = """
Hey team, just wanted to inform you that the demo is kind of ready, bu there is still stuff left.
Maybe we can show the client what we have and say rest is WIP.
Let me know what u think. Thanks.
"""

composer_task = Task(
    description="""Take the following email and rewrite it into a professional and polished version.
    Expand abbreviations.

    Email:
    {original_email}
    """,
    agent=email_agent,
    expected_output="A professional email with proper formatting and content",
)

crew = Crew(
    agents=[email_agent],
    tasks=[composer_task],
    verbose=True,
)
````

````python
response = await crew.kickoff_async({"original_email": original_email})
print(response)
````
